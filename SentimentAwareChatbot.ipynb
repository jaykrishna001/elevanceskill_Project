```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = text.lower()
    words = text.split()
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words]
    return " ".join(words)
```

```python
from transformers import pipeline

sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
def detect_sentiment(text):
    result = sentiment_analyzer(text)[0]
    return result["label"]
```

```python
def sentiment_chatbot(user_input):
    cleaned_text = preprocess_text(user_input)
    sentiment = detect_sentiment(cleaned_text)

    if sentiment == "POSITIVE":
        return "I'm glad to hear that! 😊 How can I assist you further?"
    elif sentiment == "NEGATIVE":
        return "I'm sorry you're feeling this way. 😔 Let's try to solve it together."
    else:
        return "I understand. Please tell me more."
```

```python
print(sentiment_chatbot("I am very happy with this training project"))
print(sentiment_chatbot("This course is confusing and stressful"))
print(sentiment_chatbot("Today is just another day"))
```
