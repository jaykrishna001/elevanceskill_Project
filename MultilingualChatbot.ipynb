```python
from langdetect import detect
from googletrans import Translator
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = text.lower()
    words = text.split()
    words = [lemmatizer.lemmatize(w) for w in words if w not in stop_words]
    return " ".join(words)
```

```python
translator = Translator()

def detect_language(text):
    return detect(text)

def translate_to_english(text, lang):
    if lang != 'en':
        return translator.translate(text, src=lang, dest='en').text
    return text

def translate_from_english(text, lang):
    if lang != 'en':
        return translator.translate(text, src='en', dest=lang).text
    return text
```

```python
def multilingual_chatbot(user_input):
    user_lang = detect_language(user_input)

    english_text = translate_to_english(user_input, user_lang)
    processed_text = preprocess_text(english_text)

    response_en = f"You said: {processed_text}. How can I help you further?"

    final_response = translate_from_english(response_en, user_lang)
    return final_response
```

```python
print(multilingual_chatbot("I love this project"))
print(multilingual_chatbot("मुझे यह प्रोजेक्ट पसंद है"))
print(multilingual_chatbot("J'aime ce projet"))
```
